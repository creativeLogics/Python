```python
# Question 5:
# Write a code in python in which create a class named it Car which have 5 attributes such like (model, color and name etc.)
# and 3 methods. And create 5 object instance from that class.
# Answer:
class Car:
    def __init__(self, name, color, brand):
        self.name= name
        self.model= "2019"
        self.color= color
        self.brand= brand
        self.transmition= "manual";
        
    def checkSpec(self):
        print("Car Name:", self.name,"\n")
        print("Car model:", self.model,"\n")
        print('Color:', self.color,"\n")
        print("Brand name:", self.brand,"\n")
        print("Transmition", self.transmition, "\n")
    
    def updateAndCheckSpec(self, model, transmition):
        print("Specification Changed \n")
        self.model= model
        self.transmition= transmition
        self.checkSpec()
        
    def changeAndCheckBrand(self, brand):
        print("Brand Changed \n")
        self.brand= brand
        self.checkSpec()
        
print("\n         My First Car \n")
car1= Car('Mehran', 'Black', 'Suzuki')
car1.checkSpec()
car1.updateAndCheckSpec("2020", "Auto")
car1.changeAndCheckBrand("Nissan")

print("\n         My Second Car \n")
car2= Car('Aqua', 'Blue', 'Toyota')
car2.checkSpec()
car2.updateAndCheckSpec("2020", "Auto")
car2.changeAndCheckBrand("Suzuki")

print("\n         My Third Car \n")
car3= Car('Alto', 'Red', 'Suzuki')
car3.checkSpec()
car3.updateAndCheckSpec("2020", "Auto")
car3.changeAndCheckBrand("Toyota")

print("\n         My Fourth Car \n")
car4= Car('Juke', 'Golden', 'Nissan')
car4.checkSpec()
car4.updateAndCheckSpec("2020", "Auto")
car4.changeAndCheckBrand("Kia")

print("\n         My Fifth Car \n")
car5= Car('Corolla', 'Silver', 'Toyota')
car5.checkSpec()
car5.updateAndCheckSpec("2020", "Auto")
car5.changeAndCheckBrand("Nissan")

print("Insha Allah In future")
```
